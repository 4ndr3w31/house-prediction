```python
x = df[['number of bedrooms','number of bathrooms','living area','condition of the house','Number of schools nearby', 'Distance from the airport']]
y = df[["Price"]]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

LINEAR REGRESSION

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predslr = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, predslr)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, predslr)
```

Decision Tree

```python
param_grid = {
    'criterion': ['squared_error', 'absolute_error', 'poisson'],
    'max_depth': [None, 10, 20,],
    'splitter' : ["best","random"],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
grid_tree.fit(x_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(x_test)
```

```python
mean_squared_error(y_test, tree_preds)
```

RANDOM FOREST REGRESSOR

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    'n_estimators': [2,3,5,7,10],
    'max_depth': [None, 10, 20,],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
gridrfr = GridSearchCV(estimator = rfrmodel, param_grid = param_gridrfr)
```

```python
gridrfr.fit(x_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(x_test)
```

```python
mean_squared_error(y_test, rfrpredictions)
```

```python
mean_absolute_error(y_test, rfrpredictions)
```

```python
import joblib
joblib.dump(gridrfr,'model.pkl')
```
